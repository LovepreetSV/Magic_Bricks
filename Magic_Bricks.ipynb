### Distribution of prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

### Price vs Area

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area', y='Price', data=df)
plt.title('Price vs Area')
plt.xlabel('Area (sqft)')
plt.ylabel('Price')
plt.show()
```

### Average price per BHK

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='BHK', y='Price', data=df, estimator=sum)
plt.title('Total Price per BHK')
plt.xlabel('BHK')
plt.ylabel('Total Price')
plt.show()
```

### Average price by Location

```python
plt.figure(figsize=(12, 8))
location_price = df.groupby('Location')['Price'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=location_price.values, y=location_price.index)
plt.title('Average Price by Top 10 Locations')
plt.xlabel('Average Price')
plt.ylabel('Location')
plt.show()
```

### Count plot of Parking availability

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Parking', data=df)
plt.title('Parking Availability')
plt.xlabel('Parking')
plt.ylabel('Count')
plt.show()
```

### Transaction types

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Transaction', data=df)
plt.title('Distribution of Transaction Types')
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.show()
```

### Box plot of price per square foot by District

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='District', y='Per_Sqft', data=df)
plt.title('Price per Square Foot by District')
plt.xlabel('District')
plt.ylabel('Price per Sqft')
plt.xticks(rotation=90)
plt.show()
```
